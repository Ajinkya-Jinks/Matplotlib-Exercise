
# Given Solution

```python
import pandas as pd
import matplotlib.pyplot as plt  

df = pd.read_csv("company_sales_data.csv")
profitList = df ['total_profit'].tolist()
monthList  = df ['month_number'].tolist()
plt.plot(monthList, profitList, label = 'Month-wise Profit data of last year')
plt.xlabel('Month number')
plt.ylabel('Profit in dollar')
plt.xticks(monthList)
plt.title('Company profit per month')
plt.yticks([100000, 200000, 300000, 400000, 500000])
plt.show()
```

# Exercise Question 2: Get Total profit of all months and show line plot with the following Style properties

```python
plt.figure(figsize=(9,6))
x = data["month_number"].tolist()
y = data["total_profit"].tolist()
plt.plot(x,y, label = "Profit Data of Last Year",color = "red",linestyle = "--",linewidth = 3,
         marker = "o", markerfacecolor = "k" )
plt.xticks(x)
plt.yticks([100000,200000,300000,400000,500000])
plt.xlabel("Month Number")
plt.ylabel("Profit in Dollar")
plt.legend(loc = "lower right")
plt.show()
```

# Exercise Question 3: Read all product sales data and show it  using a multiline plot

```python
x = data.month_number.tolist()

plt.figure(figsize=(9,6))
plt.plot(x,data.facecream.tolist(),label="Face Cream Sales Data",color = "blue",marker = "o",linewidth = 3)
plt.plot(x,data.facewash.tolist(),label="Facewash Sales Data",color = "orange",marker = "o", linewidth = 3)
plt.plot(x,data.toothpaste.tolist(), label = "Toothpaste Sales Data", color = "green", marker = "o", linewidth = 3)
plt.plot(x,data.bathingsoap.tolist(), label ="Bathing Soap Sales Data", color = "#134534", marker = 'o', linewidth = 3)
plt.plot(x,data.shampoo.tolist(),label = "Shampoo Sales Data", color = "#134939",marker = "o", linewidth = 3)
plt.plot(x,data.moisturizer.tolist(), label = "Moisturizers Sales Data", color = "red", marker = "o", linewidth = 3)
plt.xticks(x)
plt.yticks([1000,2000,4000,6000,8000,10000,12000,15000,18000])
plt.legend(loc = "upper left")
plt.xlabel("Month Number")
plt.ylabel("Sales Units in Number")
plt.title("Sales Data")
plt.show()
```

# Exercise Question 4: Read toothpaste sales data of each month and show it using a scatter plot

```python
plt.figure(figsize=(9,6))
x = data.month_number.tolist()
y = data.toothpaste.tolist()
plt.scatter(x,y,color = "Blue", label = "Toothpaste Sales")
plt.grid(True, linewidth = 1, linestyle = "--")
plt.xlabel("Month Number")
plt.ylabel("Salels Amount")
plt.title("Sales Data of Toothpaste")
plt.xticks(x)
plt.yticks(list(range(4500,9500,500)))
plt.legend(loc = "upper left")
plt.show()
```

# Exercise Question 5: Read face cream and facewash product sales data and show it using the bar chart


```python
plt.figure(figsize = (9,6))
x = data["month_number"].tolist()
plt.bar([a-0.25 for a in x],data.facewash.tolist(), label = "Face Wash Sales", color = "orange",width = 0.25,align = 'edge')
plt.bar([a+0.25 for a in x],data.facecream.tolist(),label = "Face Cream Sales",color = "blue",width = -0.25,align = 'edge')
plt.grid(True, linestyle = "--", linewidth = 1)
plt.xlabel("Month Number")
plt.ylabel("Sales Units in Number")
plt.legend(loc = "upper left")
plt.xticks(x)
plt.show()
```

# Exercise Question 7: Read the total profit of each month and show it using the histogram to see most common profit ranges

# Exercise Question 8: Calculate total sale data for last year for each product and show it using a Pie chart

```python
# Loop for doing sum of columns
sales_data = []
for i in range(1,7):
    add = data.iloc[i].sum()
    sales_data.append(add)
print(sales_data)

labels = ['FaceCream', 'FaseWash', 'ToothPaste', 'Bathing soap', 'Shampoo', 'Moisturizer']
plt.figure(figsize=(9,6))
plt.pie(sales_data,labels = labels, autopct = "%1.1f%%")
plt.title("Sales Data")
plt.legend(loc = "upper right")
plt.show()
```

# Exercise Question 9: Read Bathing soap facewash of all months and display it using the Subplot

```python
plt.figure(figsize=(9,6))
f,axarr = plt.subplots(2, sharex = True)
axarr[0].plot(data.month_number.tolist(),data.bathingsoap.tolist(), label = "Bathing Soap Sales", color = "#433666", marker = "o")
axarr[0].set_title("Bathing Soap Sales")
axarr[1].plot(data.month_number.tolist(),data.facewash.tolist(), label = "Facewash Sales",color = '#122686', marker = "o" )
axarr[1].set_title("Facewash Sales")
plt.show()
```

# Exercise Question 10: Read all product sales data and show it using the stack plot 

```python
# Solution given on web
df = pd.read_csv("company_sales_data.csv")
monthList  = df ['month_number'].tolist()

faceCremSalesData   = df ['facecream'].tolist()
faceWashSalesData   = df ['facewash'].tolist()
toothPasteSalesData = df ['toothpaste'].tolist()
bathingsoapSalesData   = df ['bathingsoap'].tolist()
shampooSalesData   = df ['shampoo'].tolist()
moisturizerSalesData = df ['moisturizer'].tolist()

# This code is used only for label/legends
# plt.plot([],[],color='m', label='face Cream', linewidth=5)
# plt.plot([],[],color='c', label='Face wash', linewidth=5)
# plt.plot([],[],color='r', label='Tooth paste', linewidth=5)
# plt.plot([],[],color='k', label='Bathing soap', linewidth=5)
# plt.plot([],[],color='g', label='Shampoo', linewidth=5)
# plt.plot([],[],color='y', label='Moisturizer', linewidth=5)

plt.stackplot(monthList, faceCremSalesData, faceWashSalesData, toothPasteSalesData, 
              bathingsoapSalesData, shampooSalesData, moisturizerSalesData, 
              colors=['m','c','r','k','g','y'])

plt.xlabel('Month Number')
plt.ylabel('Sales unints in Number')
plt.title('Alll product sales data using stack plot')
plt.legend(loc='upper left')
plt.show()
```
